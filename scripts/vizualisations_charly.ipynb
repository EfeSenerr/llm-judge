```python
# create dataframe for articles+summaries sentiment by humans
df_human_sentiment=df_human_article_sentiment+df_human_summary_sentiment
df_human_sentiment
```

```python
# create dataframes containing counts of each sentiment rating

df_human_sentiment_counts=df_human_sentiment.sum(axis=1)
df_LLM_sentiment_counts=df_LLM_sentiment.sum(axis=1)
df_LLM2_sentiment_counts=df_LLM2_sentiment.sum(axis=1)
```

```python

# create bar plot of proportion of sentiment scores given to both articles and summaries by different evaluators
df_evaluator_sentiment_plot = pd.DataFrame({
    'Human': (100 * df_human_sentiment_counts / df_human_sentiment_counts.sum()),
    'Llama (Prompt 1)': (100 * df_LLM_sentiment_counts / df_LLM_sentiment_counts.sum()),
    'Llama (Prompt 2)': (100 * df_LLM2_sentiment_counts / df_LLM2_sentiment_counts.sum())
})

df_evaluator_sentiment_plot.index = ['Negative', 'Neutral', 'Positive', "NA"]

ax = df_evaluator_sentiment_plot.plot(kind='bar', rot=0, color=["#A8D1D1", "#F4A261","#E76F51"])
ax.set_xlabel("Sentiment")
ax.set_ylabel("Percentage")
ax.set_title("Human vs. Llama (Prompt 1) vs. Llama (Prompt 2) Sentiment Scores")

plt.show()
```

```python
# create articles versus summaries plot

df_summary_sentiment=df_human_summary_sentiment+df_LLM_summary_sentiment+df_LLM2_summary_sentiment
df_article_sentiment=df_human_article_sentiment+df_LLM_article_sentiment+df_LLM2_article_sentiment

df_summary_sentiment_counts=df_summary_sentiment.sum(axis=1)
df_article_sentiment_counts=df_article_sentiment.sum(axis=1)

df_doctype_agg_sentiment_plot = pd.DataFrame({
    'Article': (100 * df_article_sentiment_counts / df_article_sentiment_counts.sum()),
    'Summary': (100 * df_summary_sentiment_counts / df_summary_sentiment_counts.sum())
})

df_doctype_agg_sentiment_plot.index = ['Negative', 'Neutral', 'Positive', "NA"]

ax = df_doctype_agg_sentiment_plot.plot(kind='bar', rot=0, color=["#D28A8C","#D5D1E9"])
ax.set_xlabel("Sentiment")
ax.set_ylabel("Percentage")
ax.set_title("Article vs. Summary Sentiment Scores")

plt.show()
```

```python
# create bar plot for article vs summary per evaluator

df_human_article_sentiment_counts=df_human_article_sentiment.sum(axis=1)
df_human_summary_sentiment_counts=df_human_summary_sentiment.sum(axis=1)

df_LLM_article_sentiment_counts=df_LLM_article_sentiment.sum(axis=1)
df_LLM_summary_sentiment_counts=df_LLM_summary_sentiment.sum(axis=1)

df_LLM2_article_sentiment_counts=df_LLM2_article_sentiment.sum(axis=1)
df_LLM2_summary_sentiment_counts=df_LLM2_summary_sentiment.sum(axis=1)

df_doctype_human_plot = pd.DataFrame({"Article": (100 * df_human_article_sentiment_counts / df_human_article_sentiment_counts.sum()), "Summary": (100 * df_human_summary_sentiment_counts / df_human_summary_sentiment_counts.sum())})
df_doctype_LLM_plot = pd.DataFrame({"Article": (100 * df_LLM_article_sentiment_counts / df_LLM_article_sentiment_counts.sum()), "Summary": (100 * df_LLM_summary_sentiment_counts / df_LLM_summary_sentiment_counts.sum())})
df_doctype_LLM2_plot = pd.DataFrame({"Article": (100 * df_LLM2_article_sentiment_counts / df_LLM2_article_sentiment_counts.sum()), "Summary": (100 * df_LLM2_summary_sentiment_counts / df_LLM2_summary_sentiment_counts.sum())})

df_doctype_human_plot.index = ['Negative', 'Neutral', 'Positive', "NA"]
df_doctype_LLM_plot.index = ['Negative', 'Neutral', 'Positive', "NA"]
df_doctype_LLM2_plot.index = ['Negative', 'Neutral', 'Positive', "NA"]

fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

df_doctype_human_plot.plot(kind='bar', ax=axes[0], rot=0, color=["#D28A8C","#D5D1E9"])
axes[0].set_title("Human Evaluators")
axes[0].set_xlabel("Sentiment")
axes[0].set_ylabel("Percentage")


df_doctype_LLM_plot.plot(kind='bar', ax=axes[1], rot=0, color=["#D28A8C","#D5D1E9"])
axes[1].set_title("Llama (Prompt 1)")
axes[1].set_xlabel("Sentiment")

df_doctype_LLM2_plot.plot(kind='bar', ax=axes[2], rot=0, color=["#D28A8C","#D5D1E9"])
axes[2].set_title("Llama (Prompt 2)")
axes[2].set_xlabel("Sentiment")

plt.tight_layout()
plt.show()
```
